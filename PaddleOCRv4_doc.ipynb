```python
from paddleocr import PPChatOCRv4Doc

# Step 1: Initialize the pipeline
pipeline = PPChatOCRv4Doc(
    use_doc_orientation_classify=True,
    use_doc_unwarping=True,
    use_seal_recognition=False,
    use_table_recognition=True,
)
```

```python
res_list = pipeline.visual_predict(
    input="data_cleaned.png",
    use_common_ocr=True,
    use_seal_recognition=False,
    use_table_recognition=True,
)

from pprint import pprint

visual_info = res_list[0]["visual_info"]

# Extract normal texts (like invoice number, date, client, etc.)
text_blocks = visual_info.get("normal_text_dict", {}).get("words in other_text", "")
lines = [line.strip() for line in text_blocks.split("\n") if line.strip()]
pprint(lines)
```

```python
ocr_data = res_list[0]["layout_parsing_result"]["overall_ocr_res"]

texts = ocr_data["rec_texts"]
scores = ocr_data["rec_scores"]
boxes = ocr_data["rec_boxes"]

for i, (text, score, box) in enumerate(zip(texts, scores, boxes)):
    print(f"[{i}]")
    print(f"  Text       : {text}")
    print(f"  Confidence : {score:.3f}")
    print(f"  BBox       : {box}")
```

```python
import cv2
import numpy as np
from PIL import Image

# Step 1: Load the image using OpenCV
img_path = "data_cleaned.png"
image = cv2.imread(img_path)

# Step 2: Iterate over each OCR result and draw
for text, score, box in zip(texts, scores, boxes):
    # Convert box to integer coordinates
    box = np.array(box).astype(int).reshape(-1, 2)

    # Draw bounding box (green)
    cv2.polylines(image, [box], isClosed=True, color=(0, 255, 0), thickness=2)

    # Put text and score on top-left corner of box
    x, y = box[0]
    label = f"{text} ({score:.2f})"
    cv2.putText(image, label, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX,
                0.5, (0, 0, 255), 1, cv2.LINE_AA)

# Step 3: Save the image with visualizations
output_path = "ocr_result_with_boxes.png"
cv2.imwrite(output_path, image)

print(f"Saved output image with bounding boxes: {output_path}")
```

```python


#  ⛔ We’re skipping the MLLM branch entirely ➜ no mllm_chat_bot_config needed.

# ────────────────────────────────────────────────────────────────
#  🖼️ 4. Load your image and run the OCR ➜ vector ➜ chat pipeline
# ────────────────────────────────────────────────────────────────
from paddleocr import PPChatOCRv4Doc

pipeline = PPChatOCRv4Doc()

# 4‑A. OCR, table parsing & seal recognition
visual_predict_res = pipeline.visual_predict(
    input="/content/8.jpg",       # <-- upload/mount your file & point here
    use_doc_orientation_classify=False,
    use_doc_unwarping=False,
    use_common_ocr=True,
    use_seal_recognition=True,
    use_table_recognition=True,
)

# 4‑B. Collect the visual info blocks
visual_info_list = [item["visual_info"] for item in visual_predict_res]

# 4‑C. Embed the visual text chunks with Gemini embeddings
vector_info = pipeline.build_vector(
    visual_info_list,
    flag_save_bytes_vector=True,
    retriever_config=retriever_config,
)

# 4‑D. Ask a question that’s grounded in OCR text & vector search
chat_result = pipeline.chat(
    key_list=["A/C No"],       # e.g. 驾驶室准乘人数
    visual_info=visual_info_list,
    vector_info=vector_info,
    mllm_predict_info=None,                  # ✨ No MLLM
    chat_bot_config=chat_bot_config,
    retriever_config=retriever_config,
)

print("\n🔎 Answer:", chat_result)
```
